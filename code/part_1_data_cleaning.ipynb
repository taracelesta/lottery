#### Split winning numbers into 6 seperate columns

```python
spec_chars = ["!",'"',"#","%","&","'","(",")",
              "*","+",",","-",".","/",":",";","<",
              "=",">","?","@","[","\\","]","^","_",
              "`","{","|","}","~","–", r"\\N"]
    
for char in spec_chars:
    df['winning_numbers'] = df['winning_numbers'].str.replace(char, ' ')
```
