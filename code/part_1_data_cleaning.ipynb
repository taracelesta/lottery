```python
#code inspiration from https://www.dataquest.io/blog/tutorial-add-column-pandas-dataframe-based-on-if-else-condition/

df['b_1_odd_even'] = np.where(df['ball_1_int']%2 == 0, 0,1)
df['b_2_odd_even'] = np.where(df['ball_2_int']%2 == 0, 0,1)
df['b_3_odd_even'] = np.where(df['ball_3_int']%2 == 0, 0,1)
df['b_4_odd_even'] = np.where(df['ball_4_int']%2 == 0, 0,1)
df['b_5_odd_even'] = np.where(df['ball_5_int']%2 == 0, 0,1)
df['b_6_odd_even'] = np.where(df['ball_6_int']%2 == 0, 0,1)

#1 = odd
#0 = even
```

```python
df['odd_even'] = df['b_1_odd_even'] + df['b_2_odd_even'] + df['b_3_odd_even']+ df['b_4_odd_even']+ df['b_5_odd_even']+df['b_6_odd_even']

#0= all evens
#1= 5 even/ 1 odd
#2= 4 even/2 odd
#3= 3 even/ 3 odd
#4= 2 even / 4 odd
#5= 1 even / 5 odd
#6= 6 odd
```

```python
df['b_1_high_low'] = np.where(df['ball_1_int']< 21, 0,1)
df['b_2_high_low'] = np.where(df['ball_2_int']< 21, 0,1)
df['b_3_high_low'] = np.where(df['ball_3_int']< 21, 0,1)
df['b_4_high_low'] = np.where(df['ball_4_int']< 21, 0,1)
df['b_5_high_low'] = np.where(df['ball_5_int']< 21, 0,1)
df['b_6_high_low'] = np.where(df['ball_6_int']< 21, 0,1)

#0 = 1-20
#1 = 21-40
```

```python
df['high_low'] = df['b_1_high_low'] + df['b_2_high_low'] + df['b_3_high_low']+ df['b_4_high_low']+ df['b_5_high_low']+df['b_6_high_low']

#0= all low
#1= 5 low/ 1 high
#2= 4 low/2 high
#3= 3 low/ 3 high
#4= 2 low / 4 high
#5= 1 low / 5 high
#6= 6 high
```

```python
df['b_1_IQR'] = np.where((df['ball_1_int'] >11) & (df['ball_1_int']<31), 0,1)
df['b_2_IQR'] = np.where((df['ball_2_int'] >11) & (df['ball_2_int']<31), 0,1)
df['b_3_IQR'] = np.where((df['ball_3_int'] >11) & (df['ball_3_int']<31), 0,1)
df['b_4_IQR'] = np.where((df['ball_4_int'] >11) & (df['ball_4_int']<31), 0,1)
df['b_5_IQR'] = np.where((df['ball_5_int'] >11) & (df['ball_5_int']<31), 0,1)
df['b_6_IQR'] = np.where((df['ball_6_int'] >11) & (df['ball_6_int']<31), 0,1)

#0 = 11-30
#1 = 1-10 & 31-40
```

```python
df['IQR'] = df['b_1_IQR'] + df['b_2_IQR'] + df['b_3_IQR']+ df['b_4_IQR']+ df['b_5_IQR']+df['b_6_IQR']

#0= all OQR (outer quartile range)
#1= 5 OQR/ 1 IQR
#2= 4 OQR/2 IQR
#3= 3 OQR/ 3 IQR
#4= 2 OQR / 4 IQR
#5= 1 OQR / 5 IQR
#6= 6 IQR
```
