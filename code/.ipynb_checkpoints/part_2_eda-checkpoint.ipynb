### Hot Cold Numbers

Hot numbers are the numbers that have been picked most frequently throughout the drawings. You might consider these "lucky" numbers to pick. 34 is our hottest number!

Cold numbers are drawn less frequently overall. A cold number may have been drawn recently, but the overall number of times it has been drawn is below average. You might consider these unlucky numbers. 6 is our coldest number!

People have favorite numbers, so when popular numbers come up you are sharing the winnings with lots of people. Birthdays are popular choices, so do people choose 1-12 and 1-31 more often?

If so maybe you should choose unpopular numbers so when you DO win you get more money. 

```python
#code inspiration from https://towardsdatascience.com/dealing-with-list-values-in-pandas-dataframes-a177e534f173

def to_1D(series):
    return pd.Series([x for _list in series for x in _list])
```

```python
def clean_alt_list(list_):
    list_ = list_.replace("   ", "','")
    return list_
```

```python
df['winning_numbers_new'] = df['winning_numbers'].apply(clean_alt_list)
df['winning_numbers_new'] = "'" + df['winning_numbers_new'].astype(str)
df['winning_numbers_new'] = df['winning_numbers_new'].astype(str) + "'"
```

```python
def string_to_list(list_):
    list_ = list_.replace(', ', '","').split(',')
    return list_
```

```python
df['winning_numbers_new'] = df['winning_numbers_new'].apply(string_to_list)
```

```python
def scatterplot(x_data, y_data, x_label="", y_label="", title="", color = "r", yscale_log=False):

    _, ax = plt.subplots()

    ax.scatter(x_data, y_data, s = 10, color = color, alpha = 0.75)

    if yscale_log == True:
        ax.set_yscale('log')

    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
```

```python
def plot_series(df, cols=None, title='Title', xlab=None, ylab=None):
    
    plt.figure(figsize=(18,9))
    
    for col in cols:
        
        plt.plot(df[col])

    plt.title(title, fontsize=26)
    plt.xlabel(xlab, fontsize=20)
    plt.ylabel(ylab, fontsize=20)
    
    plt.xticks(fontsize=18)
    plt.yticks(fontsize=18);
```

```python
def scatterplot(x_data, y_data, x_label="", y_label="", title="", color = "r", yscale_log=False):

    _, ax = plt.subplots()

    ax.scatter(x_data, y_data, s = 10, color = color, alpha = 0.75)

    if yscale_log == True:
        ax.set_yscale('log')

    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
```

```python
scatterplot(df['jackpot'], df['jackpot_co_winners'], x_label="jackpot prize", y_label="jackpot winners", title="Jackpot", color = "r", yscale_log=False)
scatterplot(df['match_5_prize'], df['match_5_co_winners'], x_label="match 5 prize", y_label="match 5 winners", title="Match 5", color = "r", yscale_log=False)
scatterplot(df['match_4_prize'], df['match_4_co_winners'], x_label="match 4 prize", y_label="match 4 winners", title="Match 4", color = "b", yscale_log=False)
scatterplot(df['match_3_prize'], df['match_3_co_winners'], x_label="match 3 prize", y_label="match 3 winners", title="Match 3", color = "g", yscale_log=False)
```
